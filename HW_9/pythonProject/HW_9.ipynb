```python
import pandas as pd
import nltk
nltk.download('punkt_tab')
```

```python
positive_words = ['happy', 'excited', 'awesome']
negative_words = ['disappointed', 'frustrated', 'terrible']
def rule_mood(row):
    words = nltk.word_tokenize(row['text'].lower())
    num_positive = sum([1 for word in words if word in positive_words])
    num_negative = sum([1 for word in words if word in negative_words])
    if num_positive > num_negative:
        return 2
    elif num_negative > num_positive:
        return 1
    else:
        return 0
```

```python
data_manual = data.sample(frac=0.1)
data_manual
```

```python
data_manual['Mood'] = data_manual.apply(rule_mood, axis=1)
```

```python
labeled_reviews, unlabeled_reviews = train_test_split(data, train_size=0.1, random_state=50)
```

```python
labeled_reviews['text'] = labeled_reviews['text'].astype(str)
```

```python
labeled_reviews['Mood'] = labeled_reviews.apply(rule_mood, axis=1)
unlabeled_reviews['Mood'] = unlabeled_reviews.apply(rule_mood, axis=1)
```

```python
labeled_reviews.info()
```

```python
def train_model(data):
    vect = TfidfVectorizer()
    x = vect.fit_transform(data['text'])
    y = data['Mood']
    model = LogisticRegression()
    model.fit(x,y)
    return model, vect

model, vect = train_model(labeled_reviews)
```

```python
x_unlabeled = vect.transform(unlabeled_reviews['text'])
y_unlabeled_predicted = model.predict(x_unlabeled)
```

```python
import numpy as np
```

```python
y_unlabeled_proba = model.predict_proba(x_unlabeled)
unc = -(y_unlabeled_proba * np.log2(y_unlabeled_proba)).sum(axis=1)
```

```python
unc
```

```python
y_unlabeled_proba
```

```python
labeled_reviews_new = unlabeled_reviews.iloc[unc.argsort()[:100]]
unlabeled_reviews_new = unlabeled_reviews.iloc[unc.argsort()[100:]]
```

```python
labeled_reviews = pd.concat([labeled_reviews, labeled_reviews_new])
```

```python
model, vect = train_model(labeled_reviews)
```

```python
reviews_test = unlabeled_reviews
reviews_test
```

```python
x_test = vect.transform(reviews_test['text'])
y_test_predicted = model.predict(x_test)
y_test_predicted
```

```python
f1 = f1_score(reviews_test['Mood'], y_test_predicted)
f1
```
